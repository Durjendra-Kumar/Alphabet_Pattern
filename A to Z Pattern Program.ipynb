```python
while True:

#take an alphabet from the user
    Alphabet=input("Enter your Alphabet:")

#take number of rows for the size of alphabet
    n=int(input("Enter a odd number <= 5 for rows:"))

#--------FOR PRINT "A"-------------#

    if Alphabet=="A" or Alphabet=="a":
        for i in range(n):
            for j in range(n+(n-1)):
                if (i+j==n-1) or (j-i==n-1) or (i==n//2 and j in range(n//2,n//2+(n-1))):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")
        
#--------FOR PRINT "B"-------------#    

    elif Alphabet=="B" or Alphabet=="b":
        for i in range(n):
            for j in range(n):
                if (i in range(0,n) and j==0) or (i%(n//2)==0 and j in range(0,n//2+1) or (j==n//2+1 and i%(n//2)!=0)):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#--------FOR PRINT "C"-------------#

    elif Alphabet=="C" or Alphabet=="c":
        for i in range(n):
            for j in range(n-2):
                if(i in range(1,n-1) and j==0) or (i%(n-1)==0 and j in range(1,n-3) or (j==n-3 and i==1 or i==n-2 and j==n-3)):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")
        
#---------FOR PRINT "D"-----------#

    elif Alphabet=="D" or Alphabet=="d":
        for i in range(n):
            for j in range(n):
                if(i in range(n) and j==0) or (i==0 and j in range(n-1)) or (i==n-1 and j in range(n-1)) or (j==n-1 and i in range(1,n-1)):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")
        
#--------FOR PRINT "E"-----------#

    elif Alphabet=="E" or Alphabet=="e":
        for i in range(n):
            for j in range(n):
                if(i in range(n) and j==0) or (i==0 and j in range(n)) or (i==n-1 and j in range(n)) or (i==(n//2) and j in range(n-1)):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")
        
#-----------FOR ALPHABET "F"--------#

    elif Alphabet=="F" or Alphabet=="f":
        for i in range(n):
            for j in range(n):
                if(i in range(n) and j==0) or (i==0 and j in range(n)) or (i==(n//2) and j in range(n-1)):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#----------FOR PRINT "G"----------#

    elif Alphabet=="G" or Alphabet=="g":
        for i in range(n):
            for j in range(n-1):
                if(i in range(1,n-1) and j==0) or (i==0 and j in range(1,n)) or (i==n-1 and j in range(1,n)) or (j==n-2 and i in range((n//2),n)) or (i==n//2 and j in range((n//2),n-2)):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")
        
#----------FOR PRINT "H"---------------#

    elif Alphabet=="H" or Alphabet=="h":
        for i in range(n):
            for j in range(n):
                if(j==0 or j==n-1 or i==n//2):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")
        
#-----------FOR PRINT "I"---------------#

    elif Alphabet=="I" or Alphabet=="i":
        for i in range(n):
            for j in range(n):
                if(i==0 or i==n-1 or j==n//2):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#------------FOR PRINT "J"------------#

    elif Alphabet=="J" or Alphabet=="j":
        for i in range(n):
            for j in range(n):
                if(i==0  or i in range(0,n-1) and j==n//2 or (i==n-1 and j in range(1,(n//2))) or (i in range((n//2+1),n-1) and j==0)):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#--------------FOR PRINT "K"--------------------#

    elif Alphabet=="K" or Alphabet=="k":
        for i in range(n):
            for j in range(n-n//2+1):
                if(j==0) or (i+j==n-(n//2)) or (i-j==(n-n//2)-2):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#------------FOR PRINT "L"------------------#

    elif Alphabet=="L" or Alphabet=="l":
        for i in range(n):
            for j in range(n-1):
                if(j==0 or i==n-1):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#------------FOR PRINT "M"----------------#

    elif Alphabet=="M" or Alphabet=="m":
        for i in range(n):
            for j in range(n):
                if(j==0) or (j==n-1) or (i in range(0,n//2+1) and i==j) or (i in range(0,n//2) and i+j==n-1):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#------------FOR PRINT "N"----------------#

    elif Alphabet=="N" or Alphabet=="n":
        for i in range(n):
            for j in range(n):
                if(j==0 or j==n-1 or i-j==0 ):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#-------------FOR PRINT "O"-------------------#

    elif Alphabet=="O" or Alphabet=="o":
        for i in range(n):
            for j in range(n-2):
                if(i in range(1,n-1) and j==0) or (i in range(1,n-1) and j==n-3) or (i==0 and j in range(1,n-3)) or (i==n-1 and j in range(1,n-3)):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#-------------FOR PRINT "P"----------------#

    elif Alphabet=="P" or Alphabet=="p":
        for i in range(n):
            for j in range(n-n//2):
                if(j==0) or (i==0 and j in range(0,n-n//2-1)) or(i in range(1,n//2) and j==n-n//2-1) or (i==n//2 and j in range(0,n-n//2-1)):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#--------------FOR PRINT "Q"--------------#

    elif Alphabet=="Q" or Alphabet=="q":
        for i in range(n):
            for j in range(n-1):
                if(i in range(1,n-1) and j==0 or i in range(1,n-1) and j==n-3 or i==0 and j in range(1,n-3) or i==n-1 and j in range(1,n-3) or i in range(n//2,n) and i-j==1):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")
        
#------------------FOR PRINT "R"-------------------#

    elif Alphabet=="R" or Alphabet=="r":
        for i in range(n):
            for j in range((n-n//2)+1):
                if(j==0)or(i==0 and j in range(0,((n-n//2)-1)))or(i==n//2 and j in range(0,((n-n//2)-1)))or(i in range(1,n//2)and j==n//2)or(i in range(n//2+1,n)and i-j==n//2-1):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#------------------FOR PRINT "S"----------------#

    elif Alphabet=="S" or Alphabet=="s":
        for i in range(n):
            for j in range(n):
                if i==0 and j in range(1,n)or(i==n//2 and j in range(1,n-1))or(i==n-1 and j in range(0,n-1))or(i in range(1,n//2)and j==0)or(j==n-1 and i in range(n//2+1,n-1)):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")
        
#-----------------FOR PRINT "T"------------------#

    elif Alphabet=="T" or Alphabet=="t":
        for i in range(n):
            for j in range(n):
                if(i==0 or j==n//2):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")
        
#------------------FOR PRINT "U"--------------#

    elif Alphabet=="U" or Alphabet=="u":
        for i in range(n):
            for j in range(n):
                if(i in range(0,n-1)and j==0)or(j==n-1 and i in range(0,n-1))or(i==n-1 and j in range(1,n-1)):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")
        
#------------------FOR PRINT "V"--------------#

    elif Alphabet=="V" or Alphabet=="v":
        for i in range(n):
            for j in range(n):
                if(i in range(0,n//2+1)and(j==0))or(i in range(0,n//2+1)and(j==n-1))or(i-j==n//2)or(i in range(n//2+1,n-1)and(i+j==n+(n//2)-1)):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#------------------FOR PRINT "W"--------------#

    elif Alphabet=="W" or Alphabet=="w":
        for i in range(n):
            for j in range(n):
                if(j==0 or j==n-1)or(i in range(n//2,n-1)and i+j==n-1)or(i in range(n//2,n-1)and i==j):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#------------------FOR PRINT "X"--------------#

    elif Alphabet=="X" or Alphabet=="x":
        for i in range(n):
            for j in range(n):
                if(i==j or i+j==n-1):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#------------------FOR PRINT "Y"--------------#

    elif Alphabet=="Y" or Alphabet=="y":
        for i in range(n):
            for j in range(n):
                if(i in range(0,n//2+1)and i==j)or(i in range(0,n//2+1)and i+j==n-1)or(i in range(n//2,n)and j==n//2):
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")

#------------------FOR PRINT "Z"--------------#

    elif Alphabet=="Z" or Alphabet=="z":
        for i in range(n):
            for j in range(n):
                if i==0 or i+j==n-1 or i==n-1:
                    print("*",end="")
                else:
                    print(" ",end="")
            print("")
    else:
        print(Alphabet,"is not Exist in English Alphabet!")
        
    var=input("Do you want to print another word:")
    if "0" in var:
        break
```
